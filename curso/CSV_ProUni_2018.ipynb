# Analisando os dados dos cursos do ProUni
Com base nos dados obtidos no site brasil.io vamos analisar distribuição e valores de mensalidades de todos os cursos listados no ProUni 2018

Arquivo csv -> cursos-prouni.csv

```python
# Importando as bibliotecas
import pandas as pd
```

```python
# Carregando o arquivo para o pandas
df = pd.read_csv('cursos-prouni.csv')

# Mostrando os 5 primeiros cursos
df.head()
```

```python
# Mostrando os 5 últimos cursos
df.tail()
```

```python
# Mostrando 5 cursos aleatórios
df.sample(5)
```

```python
# Qual o tamanho do conjunto de dados em termos de (linhas, colunas)
df.shape
```

```python
# Os tipos de dados de cada coluna
df.dtypes
```

```python
# Uma outra forma de visualizar o conjuto de dados
# Nesse caso podemos ver claramente as colunas que possuem valores nulos, pois a quantidade
# de linhas é menor que o tamanho do DataFrame
df.info()
```

```python
# Análise matemática das colunas numéricas
df.describe()
```

```python
# Análise dos dados qualitativos
df.describe(include='object')
```

```python
# Todas as vezes que usamos [] estamos fazendo uma seleção, um corte, em nosso conjunto de dados
# O corte mais simples é pelo nome da coluna. Nesse caso vamos trazer todos os dados de uma determinada coluna
df['uf_busca']
```

```python
# Podemos realizar operações em cima desses cortes
# Qual a quantidade de dados na coluna uf_busca (ou qual o total de cursos habilitadas no ProUni)
df['uf_busca'].count()
```

```python
# Qual a quantidade de valores repetidos na coluna uf_busca
# Ou Quantos cursos habilitadas no ProUni por UF?
df['uf_busca'].value_counts()
```

```python
# Quantas cursos habilitadas por grau
df['grau'].value_counts()
```

```python
# Quantos cursos habilitados por turno
df['turno'].value_counts()
```

```python
# Qual a mensalidade mais alta?
df['mensalidade'].max()
```

```python
# E a mensalidade mais baixa
df['mensalidade'].min()
```

```python
# Mas eu quero saber qual o curso com mensalidade mais alta?
df[df['mensalidade'] == df['mensalidade'].max()]
```

```python
# E qual o curso com mensalidade mais baixa?
df[df['mensalidade'] == df['mensalidade'].min()]
```

```python
# Outra forma de ver os mais caros!
df.nlargest(5, 'mensalidade')
```

```python
# Outra forma de ver os mais baratos!
df.nsmallest(8, 'mensalidade')
```

```python
# Qual a mensalidade mais cara em cada UF?
df.groupby('uf_busca')['mensalidade'].max()
```

```python
# E a mensalidade mais barata em cada UF?
df.groupby('uf_busca')['mensalidade'].min()
```

```python
# Vamos ordenar as mensalidades mais caras pelo valor
df.groupby('uf_busca')['mensalidade'].max().sort_values(ascending=False)
```

```python
# Mostrando a distribuição dos valores de mensalidade
%matplotlib inline
df['mensalidade'].hist(bins=100)
```

```python
# Histograma mostrando apenas os cursos que custam menos de R$2.000,00
df[df['mensalidade'] <= 2000]['mensalidade'].hist(bins=100)
```

```python
# Vamos criar um novo DataFrame apenas com os estados da Região Sul
regiao_sul = ['PR', 'SC', 'RS']
df_regiao_sul = df[df['uf_busca'].isin(regiao_sul)].copy()

df_regiao_sul.sample(5)
```

```python
# Verificando o dataframe
df_regiao_sul.info()
```

```python
# Todos os comandos anteriores podem ser aplicados ao novo DataFrame

# Vamos ordenar as mensalidades mais caras pelo valor
df_regiao_sul.groupby('uf_busca')['mensalidade'].max().sort_values(ascending=False)
```

```python
# Outra forma de ver os mais caros!
df_regiao_sul.nlargest(5, 'mensalidade')
```

```python
# Histograma mostrando apenas os cursos que custam menos de R$2.000,00
df_regiao_sul[df_regiao_sul['mensalidade'] < 2000]['mensalidade'].hist(bins=100)
```
